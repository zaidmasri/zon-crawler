```python
# import pandas as pd

import json
from pathlib import Path
import pandas as pd


# Constant json fields: Product
PRODUCT_ASIN = "asin"
PRODUCT_NAME = "name"
PRODUCT_OVERALL_RATING = "overall_rating"
PRODUCT_TOTAL_RATING_COUNT = "total_rating_count"
PRODUCT_TOTAL_REVIEWS_COUNT = "total_reviews_count"
PRODUCT_REVIEW_LIST = "review_list"
PRODUCT_FAILED_URLS = "failed_urls"

# Constant json fields: Review
REVIEW_ID = "id"
REVIEW_RATING = "rating"
REVIEW_TITLE = "title"
REVIEW_HREF = "href"
REVIEW_COUNTRY = "country"
REVIEW_DATE = "date"
REVIEW_BODY = "body"
REVIEW_VERIFIED_PURCHASE = "verified_purchase"
REVIEW_FOUND_HELPFUL = "found_helpful"
REVIEW_USERNAME = "username"
REVIEW_USERNAME_URL = "username_url"
REVIEW_IMAGES = "images"
REVIEW_VIDEOS = "videos"
REVIEW_FOUND_UNDER = "found_under"


path = Path("./data/pfw/results")

# Initialize DataFrames for the three tables
table1 = pd.DataFrame(
    columns=[PRODUCT_ASIN, "review_id", REVIEW_VERIFIED_PURCHASE, REVIEW_FOUND_HELPFUL]
)
table2 = pd.DataFrame(columns=[PRODUCT_ASIN, "review_id", REVIEW_VIDEOS, REVIEW_IMAGES])
table3 = pd.DataFrame(
    columns=[
        PRODUCT_ASIN,
        REVIEW_ID,
        REVIEW_TITLE,
        REVIEW_BODY,
        REVIEW_DATE,
        REVIEW_RATING,
        REVIEW_USERNAME,
        REVIEW_USERNAME_URL,
    ]
)

for root, dirs, files in path.walk(top_down=True, on_error=print):
    for file in files:
        try:
            # Open and read the JSON file
            with open((root / file), "r") as f:
                json_text = f.read()
                # Parse the JSON content
                json_data = json.loads(json_text)
                asin = json_data.get(PRODUCT_ASIN)
                reviews = json_data.get(PRODUCT_REVIEW_LIST, [])

                for review in reviews:
                    # Table 1: asin, review_id, verified_purchase, found_helpful
                    table1 = pd.concat(
                        [
                            table1,
                            pd.DataFrame(
                                {
                                    PRODUCT_ASIN: [asin],
                                    REVIEW_ID: [review.get(REVIEW_ID)],
                                    REVIEW_VERIFIED_PURCHASE: [
                                        review.get(REVIEW_VERIFIED_PURCHASE)
                                    ],
                                    REVIEW_FOUND_HELPFUL: [
                                        review.get(REVIEW_FOUND_HELPFUL)
                                    ],
                                }
                            ),
                        ],
                        ignore_index=True,
                    )

                    # Table 2: asin, review_id, review_videos, review_images
                    table2 = pd.concat(
                        [
                            table2,
                            pd.DataFrame(
                                {
                                    PRODUCT_ASIN: [asin],
                                    REVIEW_ID: [review.get(REVIEW_ID)],
                                    REVIEW_VIDEOS: [review.get(REVIEW_VIDEOS)],
                                    REVIEW_IMAGES: [review.get(REVIEW_IMAGES)],
                                }
                            ),
                        ],
                        ignore_index=True,
                    )

                    # Table 3: asin, review_id, review_header,
                    # review_text, date, rating, username, username_url
                    table3 = pd.concat(
                        [
                            table3,
                            pd.DataFrame(
                                {
                                    PRODUCT_ASIN: [asin],
                                    REVIEW_ID: [review.get(REVIEW_ID)],
                                    REVIEW_TITLE: [review.get(REVIEW_TITLE)],
                                    REVIEW_BODY: [review.get(REVIEW_BODY)],
                                    REVIEW_DATE: [review.get(REVIEW_DATE)],
                                    REVIEW_RATING: [review.get(REVIEW_RATING)],
                                    REVIEW_USERNAME: [review.get(REVIEW_USERNAME)],
                                    REVIEW_USERNAME_URL: [
                                        review.get(REVIEW_USERNAME_URL)
                                    ],
                                }
                            ),
                        ],
                        ignore_index=True,
                    )
        except Exception as e:
            print(f"Error processing file {file}: {e}")
# Display the resulting DataFrames
print("Table 1:")
print(table1)

print("\nTable 2:")
print(table2)

print("\nTable 3:")
print(table3)
```
